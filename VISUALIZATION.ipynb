# ***DATA VISUALIZATION***

*Visualization in Pandas is used to represent data graphically, making it easier to identify trends, patterns, and insights. It simplifies complex datasets, enhances understanding, and supports data-driven desicion-making.*

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 agg_booking=pd.read_csv("/content/fact_aggregated_bookings.csv")
 agg_booking.head(5)
```

```python
agg_booking.room_category.value_counts().plot(kind="bar")
```

```python
booking=pd.read_csv("/content/fact_bookings.csv")
```

```python
booking.booking_platform.value_counts().plot(kind="bar",color=["brown","silver"])
plt.plot()
plt.gca().set_facecolor("grey")
plt.title("room_category")
plt.show()
```

```python
booking.booking_status.value_counts().plot(kind="bar",color=["blue","silver"])
plt.plot()
plt.gca().set_facecolor("black")
plt.title("room_category")
plt.show()
```

```python
booking.groupby("booking_platform")["revenue_realized"].sum().plot(kind="pie")
```
